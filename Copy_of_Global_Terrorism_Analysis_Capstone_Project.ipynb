### 1. Attack types

```python
# created a vairable attacks 
attacks = gtd ["Attacktype"].value_counts()
pd.DataFrame(attacks)
```

```python
# in general analysis we found that iraq is most attacked country so we do analyis on that

gtd['Attacktype'].groupby(by=gtd['Countrys_']).value_counts()['Iraq'].plot(kind ='bar',legend = True)
plt.title('Attacks in Iraq')
```

Observation :

* Almost half of the terrorism cases, Bombing or Explosion used as the primary attack type.

* Other than Bombing there 'Armed Assault', 'Assassination', 'Kidnapping' etc also used as famous attack types.

* There are very less cases of 'Hijacking', 'Barricade Incident' and 'Unarmed Assault' in terrorism.

### 2. Target types

```python
# on wich terroist targets most 
# here we imported plotly.express is the easy-to-use, high-level interface to Plotly, which operates on a variety of types of data and produces easy-to-style figures.

tagtype_count = gtd.Targettype.value_counts().sort_values()
#for bar chart 
fig = px.bar(x= tagtype_count.keys(),y=tagtype_count,color=tagtype_count.keys(),
             labels= {'x':'Target type','y':'Terrorism Attacks'})
# title and given colours ,size
fig.update_layout(title = {'text': 'global terrorsim -target types','font_color':'orange','font_family':'georgia',
                           'font_size':23,'x':0.45,'y':0.97})
fig.show()
#for pie chart
fig = px.pie(names = tagtype_count.keys(),values = tagtype_count,hole=0.45,height = 650)
fig.show()

```

Observation :

More than 75% of global terrorism cases attackers targeted 'Business', 'Government (General)', 'Police', 'Military', and 'Private Citizens & Property' areas.

'Abortion Related', 'Food or Water Supply', 'Maritime', 'Tourists', 'NGO', 'Telecommunication' are some areas where attackers may not like to attack

### 3. weapon types

```python
# storing Weapon counts in varible
#used plotly.express to perform analysis
weapon_count = gtd.Weapon.value_counts()

#given figure 
fig = px.bar(x= weapon_count.keys(),y=weapon_count,color =weapon_count.keys(),
             labels={'x':'weapon Type','y':'Terrorism Attacks'})
# Given title
fig.update_layout(title= {'text':'global terrorism - weapon Types'})

fig.show()
# performing pi chart 
fig= px.pie(names = weapon_count.keys(),values = weapon_count,hole= 0.45)
fig.show()


```

Attackers mostly used 'Unknown Explosive Type', 'Unknown Gun Type', 'Automatic or Semi-Automatic Rifle' that observed in almost 60% cases.

'Sticky Bomb', Molotov Cocktail/Petrol Bomb', 'Dynamite/TNT', 'Rifle/Shotgun (non-automatic)', 'Gasoline or Alcohol', 'Time Fuse', 'Pipe Bomb', 'Blunt Object', 'Letter Bomb', 'Hands, Feet, Fists', 'Pressure Trigger', 'Unknown Weapon Type', 'Rope or Other Strangling Device', 'Other Gun Type', 'Poisoning', 'Explosive', 'Suffocation' are weapons that attackers used rarely.
